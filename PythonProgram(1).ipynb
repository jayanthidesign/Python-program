
# **Project Title: Feedback Analyzer**
 String Methods and Slicing

**Scenario:**
 A content review team wants to analyse user feedback and comments. They want to check how frequently certain words appear and extract parts of the feedback for reporting.

**Tasks:**
* Ask the user to input a sentence (representing feedback).
* Count how many words are in the sentence.
* Check how many times the word “data” appears in the sentence, regardless of case.
* Extract and print the first 5 and last 5 characters of the feedback separately.
* Reverse the sentence and print it.



```python
feedback= input("Enter your feedback:")

word_count=len(feedback.split())
print("No of words:",word_count)

word_count_data=feedback.count('data')
print(word_count_data)

first_five=feedback[:5]
last_five=feedback[-5:]

print("First five characters in feedback:",first_five)
print("Last five characters in feedback:",last_five)

Reverse=feedback[::-1]
print("reverse the feedback:",Reverse)
```


# **Project Title: BMI Calculator**
Topic:Operators, Conditional Statements

**Scenario:**
 A health organization wants a tool to help users calculate their Body Mass Index (BMI) and understand their health category based on it.

**Tasks:**
- Ask the user to enter their weight (in kilograms) and height (in meters).
- Calculate their BMI using the formula BMI = weight / (height × height).
- Classify the BMI as underweight, normal, overweight, or obese using conditional statements.
- Display the result clearly.
- Ensure the weight and height entered are integer numbers.




```python

weight=int(input("Enter your weight in kg:"))
height=int(input("Enter your height in m:"))

bmi=weight/(height*height)

if bmi>=30:
  category= "obese"
elif bmi>=25:
  category="overweight"
elif bmi>=18.5:
  category="normal"
else:
  category="underweight"

print("Your bmi is:",bmi)
print("Category:",category)
```

# **Project Title: Shopping Cart System**
Topic:While Loop, else, break, continue


**Scenario:**
You are creating a simple billing system where customers can enter the prices of items they want to purchase. The program should keep accepting prices until the user decides to stop, then apply appropriate discounts before generating the final bill.

**Tasks:**
- Continuously accept item prices using a while loop.
- Stop taking input when the user enters 0 (sentinel value).
- Keep track of the total number of items and the subtotal amount.
- Ensure no negative prices are allowed (ask again without counting as an item).
- Apply discounts based on the subtotal:
> - 20% for totals ≥ 1000
> - 10% for totals ≥ 500
> - 5% for totals ≥ 200
> - 0% for totals < 200
- Display a clear bill summary including subtotal, discount amount, and final total.

```python
prices = []

while True:
    price = float(input("Enter item price: "))

    if price == 0:
        break
    elif price < 0:
        print("Invalid price.Please Enter positive value.")
        continue

    prices.append(price)


subtotal = sum(prices)
item_count = len(prices)


if subtotal >= 1000:
    discount_rate = 0.20
elif subtotal >= 500:
    discount_rate = 0.10
elif subtotal >= 200:
    discount_rate = 0.05
else:
    discount_rate = 0

discount = subtotal * discount_rate
final_total = subtotal - discount

# Output
print("\n----- BILL SUMMARY -----")
print("Items:", item_count)
print("Subtotal:", subtotal)
print("Discount:", discount)
print("Final Total:", final_total)
```


# **Project Title: Contact Directory**

List Methods and Slicing

**Scenario:**
 You have been assigned to build a contact management system where employees can store and retrieve contact details efficiently.

**Tasks:**
* Create a 2D list where each row holds a contact’s name, phone number, and email address.
* Allow the user to add a new contact.
* Remove the last item.
* Get new items from the users using input() and add those items to the list
* Sort the list in ascending order based on any values.

```python
contacts = [
    ['jay', 895344666, 'jay@gmail.com'],
    ['priya', 64738437538, 'priya@gmail.com']
]

print("Contact details:", contacts)

name = input("Enter your name: ")
phone = input("Enter your phone number: ")
email = input("Enter your email: ")

new_contact = [name, phone, email]

contacts.append(new_contact)

print("New contact added:", contacts)

removed = contacts.pop()
print("Removed contact:", removed)

contacts.sort()
print("Sorted contacts:", contacts)
```
